```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
```

```python
df = pd.read_csv("Customer-Churn-Records.csv")
df.head()
```

```python
df.shape
```

```python
X = df.loc[:,['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure','Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
Y = df.loc[:,"Exited"]
```

```python
Y.head()
```

```python
X = pd.get_dummies(X)
```

```python
X
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=48)
```

##ANN

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
ann.compile(optimizer='RMSprop', metrics=['accuracy'], loss='binary_crossentropy')
history = ann.fit(X_train,y_train,validation_data=(X_test, y_test), batch_size=42, epochs=100)

y_pred1 = ann.predict(X_test)
y_pred1 = (y_pred1 > 0.5)

acc_ann=accuracy_score(y_test, y_pred1)
acc_ann
```

```python
plt.figure(figsize=(5,5))
cm = confusion_matrix(y_test, y_pred1, labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1])
disp.plot()
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

##XGBoost

```python
xgmodel = xgb.XGBClassifier()
xgmodel.fit(X_train,y_train)
y_pred2 = xgmodel.predict(X_test)
print(classification_report(y_test,y_pred2))
acc_xgb = accuracy_score(y_test,y_pred2)
acc_xgb
```

```python
plt.figure(figsize=(5,5))
cm = confusion_matrix(y_test, y_pred2, labels=xgmodel.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=xgmodel.classes_)
disp.plot()
plt.show()
```

Random Forest

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred3 = rf.predict(X_test)
print(classification_report(y_test,y_pred3))
acc_rf=accuracy_score(y_test,y_pred3)
acc_rf
```

```python
plt.figure(figsize=(5,5))
cm = confusion_matrix(y_test, y_pred3, labels=rf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=rf.classes_)
disp.plot()
plt.show()
```

Support Vector Machine Classifier

```python
from sklearn.svm import SVC
sv = SVC(kernel='rbf')
sv.fit(X_train,y_train)
y_pred4 = sv.predict(X_test)
print(classification_report(y_test,y_pred4))
acc_svm=accuracy_score(y_test,y_pred4)
acc_svm
```

```python
plt.figure(figsize=(5,5))
cm = confusion_matrix(y_test, y_pred4, labels=sv.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=sv.classes_)
disp.plot()
plt.show()
```

##Multi-Model Fusion

```python
import numpy as np
from scipy.stats import mode

# Assuming you have predictions from different models: y_pred1, y_pred2, y_pred3, y_pred4

# Stack the predictions horizontally using np.column_stack
combined_predictions = np.column_stack((y_pred1, y_pred2, y_pred3, y_pred4))

# Calculate the mean along the specified axis (axis=1)
final_pred = np.mean(combined_predictions, axis=1)

# Convert the final predictions to 1D array
final_pred = (final_pred >= 0.5).astype(int)

# Evaluate the predictions
print(classification_report(y_test, final_pred))
accuracy = accuracy_score(y_test, final_pred)
print("Accuracy of multi-model fusion using mean:", accuracy)
```

```python
plt.figure(figsize=(5,5))
CM = confusion_matrix(y_test, final_pred, labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=CM,display_labels=[0,1])
disp.plot()
plt.show()
```

```python
x1 = ['ANN','XGBoost','RF','SVM','Multi-model']
y1 = [acc_ann,acc_xgb,acc_rf,acc_svm,accuracy]
data = {'Models': x1 , 'Accuracy':y1}
dfacc = pd.DataFrame(data)


plt.figure(figsize=(4,4))

sns.barplot(x='Models', y='Accuracy', data=dfacc)
for i in range(len(y1)):
        plt.annotate(str(y1[i]), xy=(i, y1[i]), ha='center', va='bottom')

plt.title('Accuracy Comparison of Machine Learning Models')

plt.show()

```
